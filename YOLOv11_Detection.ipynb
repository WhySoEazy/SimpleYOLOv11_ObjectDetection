```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")

# Train the model
results = model.train(data="/content/MR3_detect-1/data.yaml", epochs=5, imgsz=640)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load trained model
model = YOLO("/content/runs/detect/train3/weights/best.pt")

# Run inference on test images
results = model.predict(
    imgsz=1028,
    source="/content/keo-the-bac-ha-fres-barley-mint-candy-hu-600g.jpg",  # Update path
    conf=0.25,
    save=True,
)

# Print and visualize results
for result in results:
    print(f"Image: {result.path}")
    print("Detections:")
    for box in result.boxes:
        label = result.names[int(box.cls)]
        conf = box.conf.item()
        xyxy = box.xyxy[0].tolist()
        print(f" - Class: {label}, Confidence: {conf:.2f}, Box: {xyxy}")

    # Visualize
    img = result.plot()  # Annotated image
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(f"Predictions for {result.path}")
    plt.show()
```
